```python
#01.Write a Program to Calculate the Sum of Natural Numbers ?


#Example : 
#Enter Start Number : 1
#Enter End Number : 10

#-----------------
#total sum of 1 to 10 = 55
#----------------- 

start=int(input("Enter Start Number : "))
end=int(input("Enter End Number : "))
total_sum=0
for num in range(start,end+1):
    total_sum+=num
print(f"total sum of {start} to {end} = {total_sum}")
```

```python
#03.Write a Program to Display count of Alphabets in to String

   #Enter Word : p4n.in

   #total = count of Alphabets : 6
   
word=input("Enter Word : ")
count=0
for char in word:
    if char.isalpha():
        count+=1
print(f"total = count of Alphabets : {count}")
```

```python
#04.Write a Program to Count Number of Digits in an Integer

   #Enter Number : 4567
   #Count Number : total Digits : 4 
   
number=int(input("Enter Number : "))
count=0
while number > 0:
    number //= 10
    count += 1
print(f"Count Number : total Digits : {count}")
```

```python
#05.Write a Program to Reverse a Number

   #Enter Number : 4569
   #Reverse Number : 9654 
   
number=int(input("Enter Number : "))
reverse=0
while number > 0:
    digit = number % 10
    reverse = reverse * 10 + digit
    number //= 10
print(f"Reverse Number : {reverse}")
```

```python
#08.Countdown Timer:

#Write a program that takes an integer as input and prints a countdown from that number to 1 using a while loop.

number = int(input("Enter a number for countdown: "))
while number > 0:
    print(number)
    number -= 1
print("Countdown finished!")
```

```python
#10.Even Number Generator:

#Write a program that generates and prints the first 5 even numbers starting from 1 using a while loop.

count = 0
number = 1
while count < 5:
    if number % 2 == 0:
        print(number)
        count += 1
    number += 1
```

```python
#11.Factorial Calculator:

#Write a program that takes an integer as input and calculates its factorial starting from 1 using a while loop.
#Print the result.

number = int(input("Enter a number to calculate its factorial: "))
factorial = 1
i = 1
while i <= number:
    factorial *= i
    i += 1
print(f"Factorial of {number} is: {factorial}")
```
